```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.normal(5.0,1.0,200) # (mean, std. deviation, N)
m = 3
b = 60
noise=np.random.normal(5,1.0,200)
y = m * x + b + noise
```

```python
plt.scatter(x,y)
```

```python
[m_fit, b_fit] = np.polyfit(x, y, 1)
print(b_fit, m_fit)
```

```python
y_fit = m_fit * x + b_fit
```

```python
plt.scatter(x, y)
plt.plot(x, y_fit, 'r-')
```

```python
x_rand = np.random.normal(5.0,1.0,200) # (mean, std. deviation, N)
y_rand = np.random.normal(5.0,1.0,200)
```

```python
plt.scatter(x_rand, y_rand)
```

```python
m_fit_rand = 0
b_fit_rand = 0

[m_fit_rand, b_fit_rand] = np.polyfit(x_rand, y_rand, 1)
print(b_fit_rand, m_fit_rand)
```

```python
y_fit_rand = m_fit_rand * x_rand + b_fit_rand
```

```python
plt.scatter(x_rand, y_rand)
plt.plot(x_rand, y_fit_rand, 'r-')
```

```python
y_res = y_fit_rand - y_rand
r_squared = 1 - (np.var(y_res)/np.var(y_rand))
r_squared
```

```python
from scipy.stats import linregress
```

```python
slope, intercept, r_value, p_value, std_err = linregress(x, y)

slope, intercept, r_value, p_value, std_err
```

```python
plt.scatter(x, y)
plt.plot(x, intercept + slope*x, 'r')
plt.legend(['fitted','original'])
```
