```python
#Annual Mean
ts1_ann = ts1.mean(dim='time')
ts2_ann = ts2.mean(dim='time')
```

```python
#-- Regrid using ESMF -- 
#-- Create sudo 1deg lat/lon --

dstgrid = xr.Dataset({'lat': (['lat'], np.linspace(-90,90,181)),
                     'lon': (['lon'], np.linspace(0,359,360)),})
#-- regridder -- 
regridder1 = xe.Regridder(ts1_ann, dstgrid, 'bilinear', periodic= True)
regridder2 = xe.Regridder(ts2_ann, dstgrid, 'bilinear', periodic= True)

ts1_reg = regridder1(ts1_ann)
ts2_reg = regridder2(ts2_ann)
```

```python
ts_reg_anom = ts2_reg - ts1_reg
```

```python
fig = plt.figure(figsize=(11,8.5))

data=ts_reg_anom
data, lons = add_cyclic_point(data, coord=ts_reg_anom['lon'])
#divnorm=colors.DivergingNorm(vcenter=0)
ts_levels = np.arange(-5, 6, 1)

# Set the axes using the specified map projection
ax=plt.axes(projection=ccrs.PlateCarree())

# Make a filled contour plot
P1 =ax.contourf(lons, ts_reg_anom.lat, data,
            transform = ccrs.PlateCarree(),cmap='RdBu_r',levels=ts_levels,extend='both')

# Add coastlines
ax.coastlines() 

cbar = plt.colorbar(P1,shrink=0.7,orientation='horizontal',
                    label='Surface Air Temperature (K)')
```
